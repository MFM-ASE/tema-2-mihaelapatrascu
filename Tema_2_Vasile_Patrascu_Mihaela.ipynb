### Calculul randamentelor simple zilnice si a log-randamentelor zilnice

```python
#defining a function which calculates the daily simple returns
def simple_return(close_prices_list):
    simple_return = close_prices_list.pct_change(1) #1 for one day lookback
    return simple_return

#defining a function which calculates the daily log-returns
def log_return(close_prices_list):
    simple_return = close_prices_list.pct_change(1) #1 for one day lookback
    log_return = np.log(1+simple_return)
    return log_return

#calculating the daily simple returns
elrond_simple_return = simple_return(df_elrond['Close'])
uipath_simple_return = simple_return(df_uipath['Close'])

#calculating the daily log-returns
elrond_log_return = log_return(df_elrond['Close'])
uipath_log_return = log_return(df_uipath['Close'])

#dropping the NaN values
elrond_simple_return.dropna(inplace = True)
uipath_simple_return.dropna(inplace = True)

elrond_log_return.dropna(inplace = True)
uipath_log_return.dropna(inplace = True)
```

## Ipoteza de piata eficienta

O piață eficientă poate fi clasificată în 3 forme: piață eficientă în formă slabă, eficientă în formă semi-tare, eficientă în formă tare.

Pentru a demonstra că o piață eficientă este în formă slabă trebuie testată ipoteza că prețul de tranzacționare urmează un model de tip random walk. Din punct de vedere teoretic dacă prețul de tranzacționare în formă logaritmată este random walk, atunci seria randamentelor are următoarele proprietăți: este serie staționară și este un proces de tip zgomot alb.

În acest caz, vom folosi logaritmul prețului de închidere pentru a afla dacă seria este random walk și seria randamentelor este white noise.

## Testarea ipotezei de piata eficienta in forma slaba - Elrond

### Testul ADF

```python
from arch.unitroot import ADF
adf = ADF(np.log(df_elrond['Close']))
print(adf.summary().as_text())
```

Pentru a testa staționaritatea am aplicat în primă fază testul ADF (Augmented Dickey Fuller) pe seria logaritmată a prețurilor de inchidere ale criptomonedei Elrond. Conform rezultatelor putem confirma că seria este nestaționară întrucât valoarea p-value (0.104) este mai mare de pragul 0.05. Cu alte cuvinte, prețul urmează un model de tip random walk, prin urmare și seria randamentelor este un proces de tip zgomot alb. Putem confirma prin intermediul acestui test ipoteza de piață eficientă. Totuși nu este suficient să susținem aceasta ipoteză doar prin aplicarea testului ADF, ci este necesară si utilizarea altor teste, asa cum se poate observa mai jos.

### Testul KPSS

```python
from arch.unitroot import KPSS
series = np.log(df_elrond['Close'])

kpss = KPSS(series)
print(kpss.summary().as_text())
```

Se observa ca valoarea lui p-value se situează sub pragul 0.05 iar valorile statisticii KPSS (multiplicatorul Lagrange LM) sunt superioare valorilor critice pentru toate cele 3 niveluri de semnificație (1%, 5% și 10%), deci respingem ipoteza nulă a testului conform căreia procesul este staționar, confirmându-se nestaționaritatea seriei prețurilor logaritmate ale criptomonedei.

```python
from arch.unitroot import VarianceRatio
import scipy.stats as stat
import scipy
```

### Individual Variance Ratio Tests

```python
x = np.asarray(df_elrond['Close'])
x = np.log(x)
N=len(x)
vr_1=[]
vrt = []
w=[]
varvrt = []
zvrt = []
q = []
lcl=[]
ucl=[]
vr_1=[]
p_value=[]
stderr=[]
alpha=0.05
k=5

def VRTest():
    for ii in range (1, k+1):
        a=2**ii
        q.append(a)
        vr = VarianceRatio(x, a)
        vrt.append(vr.vr)
        se=np.sqrt(vr._stat_variance)/np.sqrt(vr._nobs-1)
        stderr.append(se)
        lower=vr.vr-stat.norm.ppf(1-alpha/2)*se
        upper=vr.vr+stat.norm.ppf(1-alpha/2)*se
        one=1
        zvrt.append(vr.stat)
        lcl.append(lower)
        ucl.append(upper)
        vr_1.append(one)
        p_value.append(vr.pvalue)

    return vrt,p_value,zvrt,q,stderr 

VRTest()
results=pd.DataFrame(columns=['q','VR test', 'Std. Error','z statistic', 'P-value'])
results['q']=q
results['VR test']=vrt
results['Std. Error']=stderr
results['z statistic']=zvrt
results['P-value']=p_value
results
```

În cazul testului Variance Ratio propus de Lo și MacKinley (1998) se testează ipoteza nulă conform căreia putem vorbi de existența unui proces random walk ce presupune homoscedasticitate. În tabelul de mai sus se pot observa valorile statisticii z și p-Value ale acestui test pentru valorile individuale. 

În continuare vom aplica Multiple Variance Ratio Test-Studentized Maximum Modulus Test și Multiple Variance Ratio Test-Wald Test pentru a putea compara valorile testelor statistice cu valorile critice. 

### Multiple Variance Ratio Test - Studentized Maximum Modulus

```python
z=abs(results["z statistic"]).max()
alpha_star=1-(1-alpha)**(1/k)
alpha_star
z_star=stat.norm.ppf(1-alpha_star/2)
z_star

mvr=pd.DataFrame(columns=['z Statistic','Critical z','Decision'])
mvr.at[0,'z Statistic']=z
mvr['Critical z']=z_star
if z<z_star:
    mvr['Decision']="Cannot reject the null hypothesis of random walk"
else:
    mvr['Decision']="Reject the null hypothesis of random walk"
mvr
```

Rezultatele testului Studentized Maximum Modulus infirma existența modelului de random walk, întrucât se respinge ipoteza nulă. Prin urmare nu se respectă ipoteza de piață eficientă în formă slabă.

### Multiple Variance Ratio Test - Wald Test

```python
cov1=np.zeros((k,k))
q=np.asarray(q)

for i in range (0,k-1):
    for j in range (i+1,k):
        cov1[i][j]=2*(3*q[j]-q[i]-1)*(q[i]-1)/(3*q[j])
cov2=np.transpose(cov1)
cov=cov1+cov2

for i in range (0,k):
     cov[i][i]=2*(2*q[i]-1)*(q[i]-1)/(3*q[i])
    
Wald_Test=np.matmul(np.asarray(vrt)-1,np.linalg.inv(cov))            
Wald_Test=N*np.matmul(Wald_Test,np.transpose(np.asarray(vrt)-1))
p_val= 1-stat.chi2.cdf(Wald_Test,k)
mvrw=pd.DataFrame(columns=['Wald Test','Critical Chi2','P-value','Decision'])
mvrw.at[0,'Wald Test']=Wald_Test
mvrw['Critical Chi2']=stat.chi2.ppf(alpha/2,k)
mvrw['P-value']=p_val
if p_val>0.05:
    mvrw['Decision']="Cannot reject the null hypothesis of random walk"
else:
    mvrw['Decision']="Reject the null hypothesis of random walk"
mvrw
```

 Conform testului Wald, criptomoneda Elrond nu respectă ipoteza de piață eficientă în formă slabă pentru perioada de timp analizata.

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

acorr_ljungbox(elrond_log_return, lags=[12], return_df=True)
```

În final am aplicat testul autocorelării Ljung Box pentru a testa independența rentabilităților criptomonedei Elrond. Pentru a demonstra eficiența informațională a pieței de capital, nu este suficient ca seria să fie nestaționara. Este necesar și ca rentabilitățile indicelui din perioada analizată să fie independente/necorelate pentru a justifica imposibilitatea investitorilor de a previziona evoluția viitoare a pieței și a obține câștiguri excedentare.

Ipoteza nulă a testului Ljung Box indică o necorelare a rentabilităților, implicit o imposibilitate de previziune a cotațiilor viitoare, iar pentru a confirma ipoteza eficienței în formă slabă este necesar ca această ipoteză să nu fie respinsă. 

Analizând rezultatele testului din tabelul de mai sus, concluzionăm că randamentele sunt de fapt corelate, iar piața nu este eficientă în formă slabă.

## Cum s-ar putea testa ipoteza de eficiență în forma semi-tare?

Forma semi-tare a ipotezei EMH stipulează imposibilitatea investitorilor de a obține rentabilități anormale utilizând informațiile disponibile publicului larg.

Modalitățile de studiere a prezenței acestei forme de eficiență pe piața de capital sunt variate, spre exemplu analizele macroeconomice, studiile eveniment sau analize fundamentale. Analiza fundamentală utilizează toate informațiile despre firme disponibile publicului, precum price-earning ratio, declarația dividendelor, împărțirea acțiunilor. Pe de altă parte, analiza macroeconomică utilizează toate informațiile macroeconomice globale și interne disponibile publicului, cum ar fi PIB, rata dobânzii, rata de schimb, oferta de bani și indicele piețelor bursiere străine (Hasan și Wadud, 2015). 

Studiile eveniment sunt utilizate pe scară largă în literatura academică a finanțelor pentru a
evalua efectele unui eveniment asupra valorii unei companii. Studiul reper al lui Mackinlay (1997) ce
a oferit direcțiile metodologice pentru efectuarea unui astfel de studiu a generat numeroase replici și
completări, fiind aplicat și ulterior criticat chiar și în ceea ce privește alegerile electorale din SUA din
anul 2016 (Wolfers și Zitzewitz, 2018).
